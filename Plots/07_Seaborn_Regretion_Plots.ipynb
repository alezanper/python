```python
import pandas as pd
import seaborn as sb

people = pd.read_csv('people.csv',
                    delimiter = ';'
                   )
people.head()
```

```python
people.info()
```

```python
sb.lmplot(x = 'age', 
          y = 'salary', 
          data = people, 
          hue = 'sex',
          markers = ['o', 'x'])
```

```python
sb.lmplot(x = 'age', 
          y = 'salary', 
          data = people, 
          col = 'sex',
          row = 'project',
          aspect = 2,
          size = 4)
```
