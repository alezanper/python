```python
# pyplot is a library for plotting
import matplotlib.pyplot as pt
import numpy as np

# Creating random data
data = np.random.randn(300)

# Plotting data
pt.hist(data)
pt.show()
```

```python
# Adding edgecolor 
pt.hist(data, edgecolor='black')
pt.show()
```

```python
# Creating Titles
xTitle = 'X - Axis'
yTitle = 'Y - Axis' 
title = 'Title'

# Adding to plot
pt.hist(data, edgecolor='black')
pt.xlabel(xTitle)
pt.ylabel(yTitle)
pt.title(title)

pt.show()

```
