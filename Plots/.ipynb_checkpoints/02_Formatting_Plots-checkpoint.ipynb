```python
# Formatting plot
import matplotlib.pyplot as pt
import numpy as np

# curve
x = np.linspace(0,10, 100)
y = x ** 3

# Different sizes
fig = pt.figure(figsize = (8, 2))
axe = fig.add_axes([0, 0, 1, 1])
axe.plot(x, y)

# Different sizes
fig = pt.figure(figsize = (2, 2))
axe = fig.add_axes([0, 0, 1, 1])
axe.plot(x, y)
```

```python
# Using subplots
# Formatting plot
import matplotlib.pyplot as pt
import numpy as np

# curve
x = np.linspace(0,10, 100)
y = x ** 3

# Using Subplots
fig,axe = pt.subplots(nrows = 2, ncols = 1, figsize = (8, 4))
axe[0].plot(x, y)
axe[1].plot(y, x)

pt.tight_layout()
```

```python
# Multiple lines in the same plot

# Using subplots
# Formatting plot
import matplotlib.pyplot as pt
import numpy as np

# curve
x = np.linspace(0,10, 100)

# Using legends
fig = pt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Plotting
ax.plot(x, x, label = 'line 1')
ax.plot(x, 2*x, label = 'line 2')
ax.plot(x, 3*x, label = 'line 3')

# Position
ax.legend(loc = 0)
```

```python
# Saving Figure
# A big dpi creates a big figure
fig.savefig('my_figure.jpg', dpi = 100)
```
