```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as pt

people = pd.read_csv('people.csv',
                     delimiter = ';')
people.head()
```

```python
#pt.figure(figsize=(6,3))
sb.set_context('notebook', # Big print, notebook, poster
               font_scale = 3)

sb.set_style('darkgrid') # whitegrid
sb.countplot(data = people,
             x='sex')
```

```python
sb.set_context('notebook')
sb.lmplot(x = 'age', 
          y = 'salary', 
          data = people, 
          hue = 'sex',
          palette = 'inferno') # coolwarm, magma
```
