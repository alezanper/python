```python
# Formatting plot
import matplotlib.pyplot as pt
import numpy as np

x = np.linspace(0, 5, 10)
y = x**3

f = pt.figure()
ax = f.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot(x, y, 
        color='#8FFC56',
        lw = 3,             # linewidth
        alpha = 0.7,        # For transparent
        marker = 'o',       # o, x, *, 1
        ms = 10,            # Markersize
        mfc = '#4F8C26',    # MarkerFaceColor
        mew = 3,            # MarkerEdgeWidth
        ls = '-')           # linestyle --, -., :, steps, 
```

```python
# Apparence of axes

import matplotlib.pyplot as pt
import numpy as np

x = np.linspace(0, 5, 10)
y = x**3

f = pt.figure()
ax = f.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot(x, y, color='#8FFC56', ls = '--')     
ax.set_xlim([0, 2])
ax.set_ylim([0, 4])
```
