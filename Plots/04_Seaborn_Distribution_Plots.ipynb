```python
import seaborn as sb
import pandas as pn

people = pn.read_csv('people.csv', delimiter=';')

people.head()
```

```python
sb.distplot(people['salary'], 
            hist_kws=dict(edgecolor="blue"),
              # definition
            kde = True) # curve
```

```python
sb.kdeplot(people['salary'])
```

```python
sb.jointplot(x='salary',
             y='age', 
             data = people,
             kind = 'kde') # hex, reg, kde
```

```python
sb.pairplot(people,
            palette = 'coolwarm',
            hue = 'sex') # Using categorical field
```

```python
sb.rugplot(people['salary'])
```
