```python
import seaborn as sb

flights = sb.load_dataset('flights')
flights.head()
```

```python
tips = sb.load_dataset('tips')
tips.head()
```

```python
tips.corr()
```

```python
sb.heatmap(tips.corr(), 
           annot = True,
           cmap = 'coolwarm')
```

```python
piv = flights.pivot_table(index = 'month',
                    columns = 'year',
                    values = 'passengers')
piv
```

```python
sb.heatmap(piv,
           linecolor = 'white',
           linewidths = '1')
```
