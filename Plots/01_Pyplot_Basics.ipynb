```python
# customizations
import matplotlib.pyplot as pt

# Simple line
x = range(0, 10000)
y = list(x)

pt.xlabel('X - Axis [Units]')
pt.ylabel('Y - Axis [Units]')
pt.title('Title')

pt.plot(x, y)
pt.show()
```

```python
# customizations
import matplotlib.pyplot as pt

# Simple line
x = range(0, 10000)
y = list(x)

pt.xlabel('X - Axis [Units]')
pt.ylabel('Y - Axis [Units]')
pt.title('Title')

pt.plot(x, y)

# customizating axis
tick_val = [2000,4000,6000, 8000, 10000]
tick_lab = ['2k','4k', '6k', '8k', '10k']

# Adapt the ticks on the x-axis
pt.xticks(tick_val, tick_lab)
pt.show()
```

```python
# Sub Plots
import matplotlib.pyplot as pt
import numpy as np

# curve
x = np.linspace(0,10, 100)
y = x ** 3

pt.subplot(1,2,1)
pt.title("figure 1")
pt.plot(x,y,'r')

pt.subplot(1,2,2)
pt.plot(y,x,'g')
pt.title("figure 2")
```

```python
# Creating plots using objects
import matplotlib.pyplot as pt
import numpy as np

# curve
x = np.linspace(0,10, 100)
y = x ** 3

canvas = pt.figure()

p1 = canvas.add_axes([0.1, 0.1, 0.8, 0.8])
p2 = canvas.add_axes([0.2, 0.5, 0.4, 0.3])

p1.plot(x,y)
p1.set_title('Big')

p2.plot(y,x)
p2.set_title('Small')
```

```python
# Creating subplots with objects
import matplotlib.pyplot as pt
import numpy as np

# curve
x = np.linspace(0,10, 100)
y = x ** 3

fig,axes = pt.subplots(nrows=1,ncols=2)
for m in axes:
    m.plot(x, y)
```
