```python
import seaborn as sb
import pandas as pn

people = pn.read_csv('people.csv', delimiter=';')
#tips = sb.load_dataset('tips')

people.head()
```

```python
sb.barplot(x='sex', y='salary', data=people)
```

```python
sb.countplot(x='sex', data = people)
```

```python
sb.boxplot(x = 'job',y = 'salary' , data=people, hue='sex')
```

```python
sb.violinplot(x = 'job',y = 'salary' , data=people, hue='sex', split = True)
```

```python
sb.stripplot(x = 'job',
             y = 'salary', 
             data = people, 
             jitter = True, 
             hue = 'sex', 
             dodge = True)
```

```python
sb.swarmplot(x = 'job',
             y = 'salary', 
             data = people,
             size = 6)
```

```python
sb.factorplot(x = 'job',
              y = 'salary', 
              data = people,
              kind = 'bar', 
              size = 7
             )
```
