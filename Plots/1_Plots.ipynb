```python
# Plots
import matplotlib.pyplot as pt

# Simple line
x = range(0, 1000)
y = list(x)

pt.plot(x, y)
pt.show()
```

```python
# Changing the line style 
import matplotlib.pyplot as pt

# Simple line
x = range(0, 1000)
y = list(x)

pt.scatter(range(0, 20), range(0, 20))
pt.show()
```

```python
# Change the scale of x-axis
import matplotlib.pyplot as pt

# Simple line
x = range(0, 10000)
y = list(x)

pt.plot(x, y)
pt.xscale('log')
pt.show()
```

```python
# customizations
import matplotlib.pyplot as pt

# Simple line
x = range(0, 10000)
y = list(x)

pt.xlabel('X - Axis [Units]')
pt.ylabel('Y - Axis [Units]')
pt.title('Title')

pt.plot(x, y)
pt.show()
```

```python
# customizations
import matplotlib.pyplot as pt

# Simple line
x = range(0, 10000)
y = list(x)

pt.xlabel('X - Axis [Units]')
pt.ylabel('Y - Axis [Units]')
pt.title('Title')

pt.plot(x, y)

# customizating axis
tick_val = [2000,4000,6000, 8000, 10000]
tick_lab = ['2k','4k', '6k', '8k', '10k']

# Adapt the ticks on the x-axis
pt.xticks(tick_val, tick_lab)
pt.show()
```
