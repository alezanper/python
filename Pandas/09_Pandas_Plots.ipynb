```python
import pandas as pd

people = pd.read_csv('people.csv',
                     delimiter = ';')

people.head()
```

```python
# Histogram 1
people['salary'].hist(bins = 30)
```

```python
# Histogram 2
people['salary'].plot(kind = 'hist', 
                      bins = 30)
```

```python
# Histogram 3
people['salary'].plot.hist()
```

```python
people.plot.scatter(x = 'age', y = 'salary', figsize = (10, 3))
```

```python
data = pd.read_csv('Data.csv',
                   delimiter = ';',
                   decimal = ',')
data
```

```python
# plot area
# needs pure number data
data.plot.area(alpha = 0.3)
```

```python
# box plot
data.plot.box()
```

```python
# plot bar
# needs pure number data
data.plot.bar()
```

```python
# stack
# needs pure number data
data.plot.bar(stacked = True)
```

```python
# kde
# needs pure number data
data['A'].plot.kde()
```

```python
# density
# needs pure number data
data.plot.density()
```
